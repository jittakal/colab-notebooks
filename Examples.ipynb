```
from matplotlib import pyplot as plt

years = range(1950, 2020, 10)
gdp = [300.2, 545.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.title("Norminal GDP")
plt.ylabel("Billions of $")
plt.show()
```

```
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West side story"]
num_oscars = [5, 11, 3, 8, 10]

xs = [i + 0.1 for i, _ in enumerate(movies)]

print(xs)

plt.bar(xs, num_oscars)
plt.ylabel('# of Academy Awards')
plt.title("My Favorite Movies")

plt.xticks([i + 0.1 for i, _ in enumerate(movies)], movies)
plt.show()
```

```
import tensorflow as tf

a = tf.constant(3.0, dtype=tf.float32)
b = tf.constant(2.0, dtype=tf.float32)
tf_add = tf.add(a, b)

with tf.Session() as session:
  print(session.run(tf_add))
  
```

```
import tensorflow as tf

x_data = 3.5
y_data = 1.5

x = tf.placeholder(dtype=tf.float32, name="x_data")
y = tf.placeholder(dtype=tf.float32, name="y_data")

tf_add = tf.add(x, y)

with tf.Session() as session:
  print(session.run(tf_add, feed_dict = {x: x_data, y: y_data}))
```

```
import tensorflow as tf
import numpy as np

x_data = np.random.rand()
mul_by = 2.0

print("Multiplication factor is ", mul_by)
print("Random input value is    ", x_data)

x = tf.placeholder(dtype=tf.float32, name="x_data")
y = tf.Variable(mul_by)

tf_mul_by = tf.multiply(x_data, mul_by)

# Add an Op to initialize global variables.
init_op = tf.global_variables_initializer()

with tf.Session() as session:
  session.run(init_op)
  output = session.run(tf_mul_by, feed_dict = {x: x_data})
  print("Out of multiplication is ", output)
```

```
# comput natual logarithm of input value using tensorflow operations

import tensorflow as tf

x_data = 1000.0

x = tf.placeholder(dtype=tf.float32)

tf_op = tf.log(x_data)

with tf.Session() as session:
  output = session.run(tf_op, feed_dict = {x: x_data})
  print(output)
```

```
# custom polynomial

import tensorflow as tf

def custom_polynomial(value):
  return (tf.subtract(3 * tf.square(value), value) + 10)

x_data = 10.0

x = tf.placeholder(dtype=tf.float32)

with tf.Session() as session:
  output = session.run(custom_polynomial(x), feed_dict= {x: x_data})
  print(output)
```
